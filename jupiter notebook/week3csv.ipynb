```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("sales.csv")
sval,eval=map(float,input("ENter the starting and ending range:").split())
nos=[0,0,0]
y=[2000,5000,10000]
print("Invoice id \t Customer type \t purchase value")
for i,x in data.iterrows():
    if eval >= x["Total"] >=sval:
        print(str(x["Invoice ID"])+" \t "+str(x["Customer type"])+" \t "+str(x["Total"]))
    if 2000 > x["Total"] >1000:
        nos[0]=nos[0]+1  
    if 5000 > x["Total"] >2001:
        nos[1]=nos[1]+1
    if 10000 > x["Total"] >5001:
        nos[2]=nos[2]+1    
       
plt.plot(nos,y)
plt.xlabel('Number of customers')
plt.ylabel('Purchasing value')
```
