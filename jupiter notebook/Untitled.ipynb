```python
class A():
    def __init__(self,name,age):
        self.name=name
        self.age=age

class B():
    def __init__(self,mrks):
        self.mrks=mrks
    
class C(A,B):
    def __init__(self,n,a,m):
        A.__init__(self,n,a)
        B.__init__(self,m)
        self.s=0
        self.avg=0
    def total(self):
        self.s=sum(self.mrks)
        self.avg=self.s/len(self.mrks)
    def display(self):
        C.total(self)
        print(f"Name:{self.name}\nAge:{self.age}\nMarks:{self.mrks}\nTotal:{self.s}\nAverage:{self.avg}")
        
      
c1=C("boy",23,[12,23,34])
c1.display()
```

```python
def Palindrome(x):
    lst=[]
    while x:
        lst.append(x%10)
        x=x//10
    return lst==lst[::-1]


y=Palindrome(1221)
print(y)
```
