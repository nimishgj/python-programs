```python
col1=["protocol_" +x for x in traindf["protocol_type"].unique() ]
col2=["service_" + x for x in traindf["service"].unique()]
col3=["flag_" + x for x in traindf["flag"].unique()]
dumcol=col1+col2+col3
```

```python
data=traindf[["protocol_type","service","flag"]]
```

```python
labeledData=data.apply(sk.LabelEncoder().fit_transform)
```

```python
onehoteddata=sk.OneHotEncoder().fit_transform(labeledData)
```

```python
newDf=pd.DataFrame(onehoteddata.toarray(),columns=dumcol)
```

```python
traindf.drop("protocol_type",axis=1,inplace=True)
traindf.drop("service",axis=1,inplace=True)
traindf.drop("flag",axis=1,inplace=True)
```

```python
newdf=traindf.join(newDf)
```
